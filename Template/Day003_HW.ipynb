### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


file i/o會以字串操作
而xlmtodict讀出則可以利用dict的方式操作

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict
with open("./Data/64_72hr_CH.xml","r",encoding="utf-8") as fd:
    xml = fd.read()
    doc = dict(xmltodict.parse(xml))
    locations = (doc["cwbopendata"]["dataset"]["locations"]["location"])
    i=0
    for location in locations:
        #print(location["locationName"])
        i+=1
    print("高雄市總共有",len(locations),"個地區有溫度資料")
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
import xmltodict
with open("./Data/64_72hr_CH.xml","r",encoding="utf-8") as fd:
    xml = fd.read()
    doc = dict(xmltodict.parse(xml))
    locations = (doc["cwbopendata"]["dataset"]["locations"]["location"])
    for location in locations:
        print(location["locationName"])
        weatherElement = location["weatherElement"]
        time = weatherElement[0]["time"]
        print(time[0]["dataTime"])
        print(time[0]["elementValue"]["measures"],":",time[0]["elementValue"]["value"])
        print("----------------------")
        
        
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

import xmltodict
with open("./Data/64_72hr_CH.xml","r",encoding="utf-8") as fd:
    xml = fd.read()
    doc = dict(xmltodict.parse(xml))
    locations = (doc["cwbopendata"]["dataset"]["locations"]["location"])
    location = locations[0]
    print(location["locationName"])
        
    times = location["weatherElement"][0]["time"]
    for time in times :
        print(time["dataTime"])
        print(time["elementValue"]["measures"],":",time["elementValue"]["value"])
    print("總共有",len(times),"筆資料")
```
