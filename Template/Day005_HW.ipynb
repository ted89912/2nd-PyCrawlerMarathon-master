## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true") # true 熱門 false最新 # pet 寵物版
response = r.text
#print(response)
data = json.loads(response)
#print(data)
#for i in data:
    #print(i["title"])
#print(type(response)) # string
#print(len(response)) #一個字算一個
print("這個API會回傳",len(data),"筆資料")
print(data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
a = 1
for i in data:
    print("第",a,"筆資料")
    a += 1
    print("標題 :",i["title"])
    print("發文時間 :",i["createdAt"])
    print("留言人數 :",i["commentCount"])
    print("案讚人數 :",i["likeCount"])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
a = 0 
t_avg_com = 0
t_avg_lik = 0
for i in data:
    a += 1
    t_avg_com +=i["commentCount"]
    t_avg_lik +=i["likeCount"]
print("熱門平均留言人數 :",t_avg_com/a)
print("熱門平均案讚人數 :",t_avg_lik/a)
```

```python
rr = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false") 
responses = rr.text
dataa = json.loads(responses)
b = 0 
f_avg_com = 0
f_avg_lik = 0
for i in dataa:
    b += 1
    f_avg_com +=i["commentCount"]
    f_avg_lik +=i["likeCount"]
print("非熱門平均留言人數 :",f_avg_com/b)
print("非熱門平均案讚人數 :",f_avg_lik/b)

```
