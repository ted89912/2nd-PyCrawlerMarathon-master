```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json

r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true") # true ç†±é–€ falseæœ€æ–° # pet å¯µç‰©ç‰ˆ
response = r.text
#print(response)
data = json.loads(response)
#print(data)
#for i in data:
    #print(i["title"])
#print(type(response)) # string
#print(len(response)) #ä¸€å€‹å­—ç®—ä¸€å€‹
print("é€™å€‹APIæœƒå›å‚³",len(data),"ç­†è³‡æ–™")
print(data[0].keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
a = 1
for i in data:
    print("ç¬¬",a,"ç­†è³‡æ–™")
    a += 1
    print("æ¨™é¡Œ :",i["title"])
    print("ç™¼æ–‡æ™‚é–“ :",i["createdAt"])
    print("ç•™è¨€äººæ•¸ :",i["commentCount"])
    print("æ¡ˆè®šäººæ•¸ :",i["likeCount"])
```
