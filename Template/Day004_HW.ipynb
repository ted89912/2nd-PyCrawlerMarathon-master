### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
import json

r =  requests.get("https://api.github.com/events")
#r.text 模擬請求動作
print(type(r.text))
j = json.loads(r.text)
print(type(j))
print(type(j[0]))
```
